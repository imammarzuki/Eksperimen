```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

prof_inv = pd.read_csv("drive/MyDrive/Seaborn/profile_invest.csv")

ord_prof = ["safe", "moderate", "aggressive"]

sns.catplot(x="profile_invest", data=prof_inv, kind="count", order=ord_prof, hue="profile_invest", palette="bright")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.catplot(x="day", y="total_bill", data=tips, kind="bar", hue="day", palette="bright")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.catplot(x="day", y="total_bill", data=tips, kind="box", whis=2.0, hue="day", palette="bright")
plt.show()
```
