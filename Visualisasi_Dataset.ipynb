```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("dataset.xlsx")
df
```

```python
nums = ['Price', 'Rating', 'Age']
cats = ['Place_Name', 'Category', 'User Location']


for i in range(len(nums)):
    plt.subplot(1, len(nums), i+1)
    sns.boxplot(y=df[nums[i]], color='blue', orient='v')
    plt.tight_layout()

```

```python
plt.figure(figsize=(30, 10))
for i in range(len(nums)):
    plt.subplot(1, len(nums), i+1)
    sns.distplot(df[nums[i]], color='blue')
    plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30, palette='bright')
plt.title('Distribusi Usia Pengguna')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 10))
# Asumsikan 'Category' adalah kolom kategorikal dalam dataframe Anda.
sns.histplot(data=df, x='Age', hue='Category', kde=True, bins=30, palette='bright')
plt.title('Distribusi Usia Pengguna Berdasarkan Kategori')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y='Place_Name', x='count', hue='count', data=df['Place_Name'].value_counts().reset_index()[:10], palette='bright')
plt.title('Jumlah Tempat Wisata dengan Rating Tertinggi', pad=20)
plt.xlabel('Jumlah Rating')
plt.ylabel('Nama Lokasi')
plt.show()
```

# Kategori Destinasi Wisata

```python
sns.countplot(y='Category', data=df, palette='bright')
plt.title('Perbandingan Jumlah Kategori Wisata di Kota Surabaya', pad=20)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df, x='Category', y='Age')
plt.title('Persebaran distribusi umur pada setiap Kategori')
plt.xlabel('Kategori')
plt.ylabel('Umur')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
category_rating = df.groupby('Category')['Rating'] \
.mean().reset_index().sort_values(by='Rating', ascending=False)

categories = category_rating['Category']
ratings = category_rating['Rating']
colors = plt.cm.tab20.colors[:len(categories)]

plt.figure(figsize=(6, 6))
plt.pie(ratings, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Rata-rata Rating Tertinggi Menurut Kategori')
plt.axis('equal')
plt.show()
```

```python
top_place_rating = df.groupby('Place_Name')['Rating'] \
.mean().reset_index().sort_values(by='Rating', ascending=False)[:10]

plt.figure(figsize=(10, 6))
sns.barplot(y='Place_Name', x='Rating', data=top_place_rating, palette='bright')
plt.title('Top 10 Rata-rata Rating Tertinggi')
plt.ylabel('Nama Kategori')
plt.xlabel('Rata-rata Rating ')
plt.show()
```

```python
top_price_place = df.groupby('Place_Name')['Price'] \
.mean().reset_index().sort_values(by='Price', ascending=False)[:10]

plt.figure(figsize=(10, 6))
sns.barplot(y='Place_Name', x='Price', data=top_price_place, palette='bright')
plt.title('Top 10 Rata-rata Harga Tertinggi')
plt.ylabel('Nama Tempat')
plt.xlabel('Rata-rata Harga')
plt.show()
```

```python
askot = df['User Location'].apply(lambda x : x.split(',')[0])

plt.figure(figsize=(8,6))
sns.countplot(y=askot, palette='bright')
plt.title('Jumlah Asal Kota dari setiap Pengunjung')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price', y='Rating')
plt.title('Korelasi antara Harga dan Penilaian')
plt.xlabel('Harga')
plt.ylabel('Penilaian')
plt.grid(True)
plt.show()
```

```python
# Misalkan kita ingin menampilkan tren untuk beberapa tempat wisata teratas berdasarkan jumlah ulasan
top_places = df['Place_Name'].value_counts().head(5).index
filtered_df = df[df['Place_Name'].isin(top_places)]
plt.figure(figsize=(12, 8))
sns.lineplot(data=filtered_df, x='User Location', y='Rating', hue='Place_Name', marker='o')
plt.title('Tren Penilaian Tempat Wisata Berdasarkan Lokasi Pengguna')
plt.xlabel('Lokasi Pengguna')
plt.ylabel('Penilaian')
plt.legend(title='Tempat Wisata')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df[nums].corr(), cmap='Blues', annot=True, fmt='.2f')
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(data=df, x='Rating', hue='Place_Name', palette='viridis')
plt.title('Frekuensi Penilaian untuk Setiap Tempat')
plt.xlabel('Penilaian')
plt.ylabel('Jumlah')
plt.legend(title='Tempat', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='Category', y='Rating', palette='Set2')
plt.title('Boxplot Penilaian Berdasarkan Kategori')
plt.xlabel('Kategori')
plt.ylabel('Penilaian')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
